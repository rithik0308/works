```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as pt
import tensorflow as tf 
import seaborn as sns

from sklearn.dummy import DummyClassifier 
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("train.csv")
train.shape
```

```python
test = pd.read_csv("test.csv")
test.shape
```

to create an base line we did dummy classification

```python
dummy_clf = DummyClassifier(strategy = "stratified")
#the reason for reassing of the train.chrun is after droping the cloumns the model should remember its target 
dummy_clf.fit(train.drop(['CustomerID','Churn'],axis=1), train.Churn)
```

```python
#why we are dropping the customer id is it will not give us any menaing full pattern or info it can infulance and derail the prediction 
predict_prob = dummy_clf.predict_proba(test.drop(["CustomerID"], axis=1))[:,1]
```

```python
predict_df = pd.DataFrame({"CustomerID ": test[["CustomerID"]].values[:,0], "prediction_prob" : predict_prob})
predict_df.shape
predict_df.head(10)
```
